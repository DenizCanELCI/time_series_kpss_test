```python
# Burada KPSS testi incelenmi≈ütir
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import kpss
```

```python
#Sƒ±caklƒ±k deƒüerlerini csv olarak import ediyoruz
path = 'https://raw.githubusercontent.com/selva86/datasets/master/daily-min-temperatures.csv'
df = pd.read_csv(path, parse_dates=['Date'], index_col='Date')
df.plot(title='Daily Temperatures', figsize=(14,8), legend=None);
```

```python
# Burada kpss testi i√ßin fonksiyon olu≈üturuyoruz
def kpss_test(series, **kw):    
    statistic, p_value, n_lags, critical_values = kpss(series, **kw)
    print(f'KPSS Statistic: {statistic}')
    # p deƒüerimiz
    print(f'p-value: {p_value}')
    # lag sayƒ±sƒ±
    print(f'num lags: {n_lags}')
    print('Critial Values:')
    for key, value in critical_values.items():
        print(f'   {key} : {value}')
    print(f'Result: The series is {"not " if p_value < 0.05 else ""}stationary')
```

```python
# kpss testindek anahtar noktamƒ±z p-value(p deƒüeri).
# Eƒüer KPSS testi sonucu hesaplanan p deƒüeri 0.05'den b√ºy√ºk ise 
# serimiz duraƒüan(stationary), aksi takdirde duraƒüan deƒüildir(non-stationary).
kpss_test(df)
```
